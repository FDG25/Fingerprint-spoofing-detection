```python
import matplotlib.pyplot as plt
import os
plot_index = 0

#print(random_numbers)

def plotDCF(x, y, labels, colors, xlabel):
    global plot_index
    plt.figure()
    for i in range(0,len(x)):
        plt.plot(x[i], y[i], label=labels[i], color=colors[i])
        plt.xlim([min(x[i]), max(x[i])])
        #plt.ylim([0,max(y[i])+1])
    plt.xscale("log")    
    plt.legend()
    plt.xlabel(xlabel)    
    plt.ylabel("min DCF")
    plt.savefig(os.path.join('prova','plot_' + str(plot_index) + '.png'))
    plot_index+=1
    plt.show()

# C and minDCF values must have the same length
C_values = [0.001,0.01,0.1,1]
minDCF1 = [0.3354,0.1235,0.2563,0.8465]

C_values2 = [0.001,0.01,0.1,1]
minDCF2 = [0.4456,0.2235,0.6055,0.8465]

C_values3 = [0.001, 0.01, 0.1, 1]
minDCF3 = [0.3354, 0.1500, 0.2750, 0.9000]

labels = ['log() = 0.5','log() = 0.8','log() = 0.9']
colors = ['b','g','r']
#b : blue.
#g : green.
#r : red.
#c : cyan.
#m : magenta.
#y : yellow.
#k : black.
#w : white
plotDCF([C_values,C_values2, C_values3],[minDCF1,minDCF2,minDCF3],labels,colors,'C')
plotDCF([C_values,C_values2, C_values3],[minDCF1,minDCF2,minDCF3],labels,colors,'C')
plotDCF([C_values,C_values2, C_values3],[minDCF1,minDCF2,minDCF3],labels,colors,'C')
plotDCF([C_values,C_values2, C_values3],[minDCF1,minDCF2,minDCF3],labels,colors,'C')
```
